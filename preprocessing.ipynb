Removing Frequent Words

```python
from collections import Counter
word_count = Counter()
for text in df['Accepted Answer']:
  for word in text.split():
    word_count[word] +=1

word_count.most_common(10)
```

```python
frequent_words = set(word for (word, wc) in word_count.most_common(3))
def remove_freq_words(text):
  return " ".join([word for word in text.split() if word not in frequent_words])
```

```python
df['Accepted Answer'] = df['Accepted Answer'].apply(lambda x: remove_freq_words(x))
df.head()
```

Removing Rare Words

```python
rare_words = set(word for (word, wc) in word_count.most_common()[:-10:-1])
rare_words
```

```python
def remove_rare_words(text):
  return " ".join([word for word in text.split() if word not in rare_words])
```

```python
df['Accepted Answer'] = df['Accepted Answer'].apply(lambda x: remove_rare_words(x))
df.head()
```

Removing Special Characters

```python
import re
def remove_spl_chars(text):
  text = re.sub('^a-zA-Z0-9', " ", text)
  text = re.sub('\s+', " ", text)
  return text
```

```python
df['Accepted Answer'] = df['Accepted Answer'].apply(lambda x: remove_spl_chars(x))
df.head(10)
```

Stemming

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
def stem_words(text):
  return " ".join([ps.stem(word) for word in text.split()])
```

Lemmatization and POS Tagging

```python
from nltk import pos_tag
from nltk.corpus import wordnet
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
wordnet_map = {"N": wordnet.NOUN, "V": wordnet.VERB, "R": wordnet.ADV}

def lemmatize_words(text):
  pos_text = pos_tag(text.split())
  return " ".join([lemmatizer.lemmatize(word, wordnet_map.get(pos[0], wordnet.NOUN)) for word, pos in pos_text])
```

```python
# nltk.download('wordnet')
# nltk.download('averaged_perceptron_tagger_eng')
```

```python
df['Accepted Answer lemmatized'] = df['Accepted Answer'].apply(lambda x: lemmatize_words(x))
df.head()
```

Removing url

```python
def remove_url(text):
  return re.sub(r'https?://\S+|www\.\S+', "", text)
```

```python
df['Accepted Answer lemmatized'] = df['Accepted Answer'].apply(lambda x: remove_url(x))
df.head()
```

Removing HTML tags

```python
def remove_html_tags(text):
  return re.sub(r"<code>.*?</code>", "", text, flags=re.DOTALL)
```

```python
df['Accepted Answer without html'] = df['Accepted Answer lemmatized'].apply(lambda x: remove_html_tags(x))
df.head()
```

```python
from spellchecker import SpellChecker
```

```python
spell = SpellChecker()
def correct_spellings(text):
  corrected_text = []
  misspelled_text = spell.unknown(text.split())
  for word in text.split():
    if word in misspelled_text:
      corrected_text.append(spell.correction(word))
    else:
      corrected_text.append(word)

  return " ".join(corrected_text)
```

```python
df['Accepted Answer Spell Corrected'] = df['Accepted Answer lemmatized'].apply(lambda x: correct_spellings(x))
df.head()
```
